Q1: Difference between a t-test and a z-test:

Z-test: A z-test is used when the population standard deviation is known, or the sample size is large (typically n > 30). It is suitable for testing hypotheses about the population mean when the population standard deviation is known.
Example scenario: Testing whether the mean height of a population is equal to a specified value when the population standard deviation is known.

T-test: A t-test is used when the population standard deviation is unknown, and the sample size is small (typically n < 30). It is used to test hypotheses about the population mean when the population standard deviation is unknown.
Example scenario: Testing whether the mean exam scores of a small sample of students are significantly different from a specified value when the population standard deviation is unknown.



Q2: Differentiate between one-tailed and two-tailed tests:

One-tailed test: In a one-tailed test, the critical region is on only one side of the distribution. It is used to test whether a population parameter is greater than or less than a specified value.
Example: Testing whether the mean exam scores of a group of students are significantly greater than a specified value.

Two-tailed test: In a two-tailed test, the critical region is split between both sides of the distribution. It is used to test whether a population parameter is not equal to a specified value.
Example: Testing whether the mean exam scores of a group of students are significantly different from a specified value.



Q3: Concept of Type 1 and Type 2 errors in hypothesis testing:

Type 1 error: Type 1 error occurs when the null hypothesis is rejected when it is actually true. It represents a false positive.
Example: Concluding that a medical treatment is effective when it actually has no effect.

Type 2 error: Type 2 error occurs when the null hypothesis is not rejected when it is actually false. It represents a false negative.
Example: Concluding that a medical treatment is ineffective when it actually has an effect.

Q8: Margin of Error and Sample Size:

The margin of error in a confidence interval represents the amount by which the sample statistic may differ from the population parameter. It is influenced by the desired confidence level, the variability in the data (standard deviation), and the sample size. A larger sample size generally leads to a smaller margin of error because larger samples provide more precise estimates of the population parameter.

Example:

Suppose you are conducting a survey to estimate the average income of households in a city. If you survey only 50 households, the margin of error might be relatively large due to the limited sample size. However, if you increase the sample size to 500 households while keeping other factors constant, the margin of error would decrease because the larger sample size provides more accurate estimates of the population parameter.
